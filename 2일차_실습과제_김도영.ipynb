```python
# 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import platform

# 운영체제를 구분하여 한글 폰트 지정
if platform.system() == 'Windows':
    font_name = fm.FontProperties(fname='../Fonts/malgun.ttf').get_name()
    plt.rc('font', family=font_name)

else:
    plt.rc('font', family='AppleGothic', size=8)

# 파일 로딩
df=pd.read_excel('subwaytime1.xls',sheet_name='지하철 시간대별 이용현황', header=[0,1])
df
```

```python
# 컬럼 확인
df.columns
```

```python
# 필요한 컬럼만 추출
subDF=df.iloc[:,[0,1,2,3,11,13]]
subDF
```

```python
# ',' 제거하고 정수형 변환 후 확인
subDF[('07:00:00~07:59:59', '하차')]=subDF[('07:00:00~07:59:59', '하차')].replace(',','',regex=True)
subDF[('08:00:00~08:59:59', '하차')]=subDF[('08:00:00~08:59:59', '하차')].replace(',','',regex=True)
subDF[('07:00:00~07:59:59', '하차')]=subDF[('07:00:00~07:59:59', '하차')].astype('int64')
subDF[('08:00:00~08:59:59', '하차')]=subDF[('08:00:00~08:59:59', '하차')].astype('int64')
subDF.info()
```

```python
# 7시~7시59분59초 하차인원 + 8시~8시59분59초 하차인원 더해서 새로운 컬럼 지정
subDF['07:00:00~08:59:59','하차']=subDF['07:00:00~07:59:59','하차']+subDF['08:00:00~08:59:59','하차']
```

```python
# 컬럼 확인
subDF
```

```python
# 값 출력 & 그래프 그리기

get_off=[]    # 최대 하차인원 담는 리스트
station=[]    # 최대 하차역 담는 리스트

for num in range(1,8):
    DF=subDF[subDF[('호선명', 'Unnamed: 1_level_1')]==str(num)+'호선']
    max_value=DF.iloc[:,-1].max()
    idx=DF.iloc[:,-1].idxmax()
    station_value=subDF.iloc[idx,3]
    print('출근 시간대 {0}호선 최대 하차역은 {1}역이고 하차 인원은 {2:,}명 입니다.'.format(num,station_value,max_value))
    get_off.append(max_value)
    station.append( f'{num}호선:{station_value}역')
    
plt.figure(dpi=100)
plt.title('출근 시간대 지하철 노선별 최대 하차 인원 및 하차역')
plt.bar(station,get_off)
plt.xticks(station, rotation=80)  # xtick 문자열을 80도 회전
plt.show() 
```

### groupby로 풀어보기

```python
# 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import platform

# 운영체제를 구분하여 한글 폰트 지정
if platform.system() == 'Windows':
    font_name = fm.FontProperties(fname='../Fonts/malgun.ttf').get_name()
    plt.rc('font', family=font_name)

else:
    plt.rc('font', family='AppleGothic', size=8)

# 파일 로딩
df=pd.read_excel('subwaytime1.xls',sheet_name='지하철 시간대별 이용현황', header=[0,1])
df
```

```python
df.columns
```

```python
df=df.iloc[:,[1,3,11,13]]
df
```

```python
df.dtypes
```

```python
# ',' 없애기
df[('07:00:00~07:59:59','하차')]=df[('07:00:00~07:59:59','하차')].apply(lambda x: x.replace(',',''))
df[('08:00:00~08:59:59','하차')]=df[('08:00:00~08:59:59','하차')].apply(lambda x: x.replace(',',''))

# object -> int 형 변환
df=df.astype({('07:00:00~07:59:59','하차'):'int64',('08:00:00~08:59:59','하차'):'int64'})
df.dtypes
```

```python
# 7시~,8시~ 더해서 7~9시 컬럼 형성
df[('07:00:00~08:59:59','하차')]=df[('07:00:00~07:59:59','하차')]+df[('08:00:00~08:59:59','하차')]
df
```

```python
# '호선명'에 대한 그룹바이
df_group=df.groupby(by=('호선명', 'Unnamed: 1_level_1'))
```

```python
# 출력 값 키, 최대값, 최대값 인덱스 확인한건데 이걸로 문제 풀면 더 쉬울 것 같지만 어떻게 하는지 모르겠습니다...
for key, group in df_group:
    print(key)
    print(group[('07:00:00~08:59:59','하차')].max())
    print(group[('07:00:00~08:59:59','하차')].argmax())
```

```python
# '1호선'에 대해 DF 어떤지 확인
df_1=df_group.get_group('1호선')
df_1
```

```python
# 최댓값, 최대값 인덱스
df_1[('07:00:00~08:59:59','하차')].max(), df_1[('07:00:00~08:59:59','하차')].argmax()
```

```python
# 최대값 인덱스 idx에 저장
idx=df_1[('07:00:00~08:59:59','하차')].argmax()
```

```python
# 최대값 인덱스에 대한 지하철역명 찾기
df_1.iloc[idx,1]
```

```python
# 값  출력 & 그래프 <- 리스트 만들기

a=[]  # 하차역 담을 리스트
b=[]  # 하차인원 담을 리스트

for num in range(1,8):
    df=df_group.get_group(str(num)+'호선')
    df[('07:00:00~08:59:59','하차')].max(), df[('07:00:00~08:59:59','하차')].argmax()
    idx=df[('07:00:00~08:59:59','하차')].argmax()
    df.iloc[idx,1]
    print('출근 시간대 {0}호선 최대 하차역은 {1}역이고 하차인원은 {2:,}명 입니다.'.format(num,df.iloc[idx,1],df[('07:00:00~08:59:59','하차')].max()))
    a.append(f'{num}호선 {df.iloc[idx,1]}역')
    b.append(df[('07:00:00~08:59:59','하차')].max())
    
plt.figure(dpi=100)
plt.bar(a, b)
plt.xticks(a, rotation=80)  # xtick 문자열을 80도 회전
plt.show()
```
