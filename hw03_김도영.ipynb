```python
# 1) 내부 조인한 결과에 ‘연락처’ 컬럼 추가
conn = pymysql.connect(host='localhost', user='root', password='0926', db='shoppingmall', charset='utf8')

cur = conn.cursor()
query='''
select u.userName, b.prodName, u.addr, concat(u.mobile1, u.mobile2) as '연락처'
from user_table as u inner join buy_table as b
on u.userID = b.userID;
'''  # 실제 쿼리와 동일한 문자열 전달 (따옴표 주의)
cur.execute(query)
rows=cur.fetchall() # 모든 데이터를 가져옴

problem1_df=pd.DataFrame(rows)
problem1_df.columns=['userName','prodName','addr','연락처']
problem1_df.index=list(range(1,13))

problem1_df
```

```python
# 2) userID가 KYM인 사람이 구매한 물건과 회원 정보 출력
conn = pymysql.connect(host='localhost', user='root', password='0926', db='shoppingmall', charset='utf8')

cur = conn.cursor()
query='''
select u.userID, u.userName, b.prodName, u.addr, concat(u.mobile1,u.mobile2)
from user_table as u inner join buy_table as b 
on u.userID = b.userID
where u.userID ='KYM';
'''  # 실제 쿼리와 동일한 문자열 전달 (따옴표 주의)
cur.execute(query)
rows=cur.fetchall() # 모든 데이터를 가져옴

problem2_df=pd.DataFrame(rows)
problem2_df.columns=['userID','userName','prodName','addr','concat(u.mobile1,u.mobile2)']
problem2_df.index=[1]

problem2_df
```

```python
# 3) 전체 회원이 구매한 목록을 회원 아이디 순으로 정렬
conn = pymysql.connect(host='localhost', user='root', password='0926', db='shoppingmall', charset='utf8')

cur = conn.cursor()
query='''
select u.userID, u.userName, b.prodName, u.addr, concat(u.mobile1,u.mobile2) as '연락'
from user_table as u inner join buy_table as b 
on u.userID = b.userID
order by u.userID;
'''  # 실제 쿼리와 동일한 문자열 전달 (따옴표 주의)
cur.execute(query)
rows=cur.fetchall() # 모든 데이터를 가져옴

problem3_df=pd.DataFrame(rows)
problem3_df.columns=['userID','userName','prodName','addr','연락']
problem3_df.index=list(range(1,13))

problem3_df
```

```python
# 4) 쇼핑몰에서 한 번이라도 구매한 기록이 있는 회원 정보를 회원 아이디 순으로 출력
# (distinct 사용)
conn = pymysql.connect(host='localhost', user='root', password='0926', db='shoppingmall', charset='utf8')

cur = conn.cursor()
query='''
select distinct u.userID, u.userName, u.addr
from user_table as u inner join buy_table as b 
on u.userID = b.userID
where b.amount >= 1
order by u.userID;
'''  # 실제 쿼리와 동일한 문자열 전달 (따옴표 주의)
cur.execute(query)
rows=cur.fetchall() # 모든 데이터를 가져옴

problem4_df=pd.DataFrame(rows)
problem4_df.columns=['userID','userName','addr']
problem4_df.index=list(range(1,6))

problem4_df
```

```python
# 5) 쇼핑몰 회원 중에서 주소가 경북과 경남인 회원 정보를 회원 아이디 순으로 출력
conn = pymysql.connect(host='localhost', user='root', password='0926', db='shoppingmall', charset='utf8')

cur = conn.cursor()
query='''
select u.userID, u.userName, u.addr, concat(u.mobile1, u.mobile2) as '연락'
from user_table as u inner join buy_table as b 
on u.userID = b.userID
where u.addr in ('경북', '경남')
order by u.userID;

'''  # 실제 쿼리와 동일한 문자열 전달 (따옴표 주의)
cur.execute(query)
rows=cur.fetchall() # 모든 데이터를 가져옴

problem5_df=pd.DataFrame(rows)
problem5_df.columns=['userID','userName','addr','연락']
problem5_df.index=list(range(1,5))

problem5_df
```
